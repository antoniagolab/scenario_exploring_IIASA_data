For this assigment, chapter 2 of the _SR15_ report by the IPCC was thouroughly read. Based on this and _IIASA data resources_, the created figures and tables will focus on one of the key findings about characteristics of 1.5° pathways: __Considerable shifts in investment patterns__ (see Table 2.5 in _SR15_) which states that investments in fossil fuels are mostly projected to decline. 

This aspect will be analyzed resistricted to the region of Asia (R5ASIA) which is the region with the highest energy consumption and the model _MESSAGEix_. 


First, the given scenarios of _MESSAGEix_ will be explored, then the investments in fossil extraction are illustrated based on the modelled scenarios which aim to achieve a temperature rise of below 2°C. At last, a brief comparison to projected investments in wind and solar energy is made.

```python
# importing all needed packages
import numpy as np
import pandas as pd
import pyam
import matplotlib.pyplot as plt
```

## __1__: Exploration of on _MESSAGEix_

```python
# Establishing connection:

conn = pyam.iiasa.Connection()
conn.connect('iamc15')
```

```python

df_asia = pyam.read_iiasa(
    'iamc15',
    model='MESSAGEix*',
    region= 'R5ASIA',meta=['category'])

df_asia.scenario
```

A central difference between these scenarios of _CD-Links_ lies in the assumptions on policy making, specifically in national policies (Source: _CD-Links Scenario Database_), some of the scenarios imply varying chances of reaching the goal of staying below 2°C based on following different "policy paths". The _NPi_ is calculated based on the implementation of policies which are/were imposed in G20 countries and are saved in a corresponding policy database.

Based on the information given in the _CD-Links Scenario Database_, the following table is constructed:

```python
scenarios = pd.DataFrame()
scenarios['name of scenario'] = df_asia.scenario[:-1]
scenarios['overview'] = ['National Policies until 2020, after which implementation of Nationally Determined Contributions (NDCs) until 2025/2030',
                        'National Policies until 2030, equivalent effort relative to the NoPolicy scenario after 2030', 'National Policies until 2020, as of 2020 staying within 1000 GtCO₂ budget for 2011-2100 period, corresponding to a >66% chance of staying below 2°C throughout 21st century', 
                        'National Policies until 2020, as of 2020 staying within 1600 GtCO₂ budget for 2011-2100 period, corresponding to a >50% chance of staying below 2°C throughout 21st century', 'National Policies until 2020, as of 2020 staying within 400 GtCO₂ budget for 2011-2100 period, corresponding to a >66% chance of staying below 1.5°C in 2100', 'Baseline scenario without climate policy']
pd.set_option('display.max_colwidth', 500)
display(scenarios)
```

```python
df_asia = pyam.read_iiasa(
    'iamc15',
    model='MESSAGEix*',
    region= 'R5ASIA', variable=['Emissions|CO2|Energy'])

df_asia.plot(legend=dict(loc='center left',bbox_to_anchor=(1.0, 0.5)))
```

Based on the illustrated CO2 emission projections in the figure above, it seems like this set of scenarios can be distinguished roughly in two groups: While scenarios _CD-Links INDCi_, _CD-Links Npi_ and _CD-Links NoPolicy_ project increasing CO2 emissions in the future, the other three ones project decreasing. This is due to the fact that _CD-Links NPi2020 1600_, _1000_ and _400_ are models projecting variables under consideration of achieving below 2°C temperature rise. The numbers _1600_, _1000_ and _400_ are implying the carbon budgets (GtCO2) limiting globale warming whithin the model. 


The following figure shows the categorization of the scenarios based on the modelled temperature rise outcome:

```python
df_asia = pyam.read_iiasa(
    'iamc15',
    model='MESSAGEix*',
    region= 'R5ASIA', variable=['Emissions|CO2|Energy'])

df_asia.plot(color= 'category',legend=dict(loc='center left',bbox_to_anchor=(1.0, 0.5)))
```

As the focus in this task lies on key characteristica of modelled scenarios with the goal of staying below 2°C, _CD-Links INDCi_, _CD-Links Npi_ and _CD-Links NoPolicy_ will further not be included.

```python
scens = ['CD-LINKS_NPi2020_400','CD-LINKS_NPi2020_1000','CD-LINKS_NPi2020_1600']
```

## __2__: Investements in extraction of fossil fuel

```python
inv_1 = 'Investment|Energy Supply|Extraction|Fossil'
df_asia = pyam.read_iiasa(
    'iamc15',
    model='MESSAGEix*',
    region= 'R5ASIA', variable=[inv_1],scenario=scens)

df_asia.plot(legend=dict(loc='center left',bbox_to_anchor=(1.0, 0.5)))
```

Timeseries data on the development of investments in fossil extraction is here only given for the two scenarios displayed above. 

For both, a peak of investements for the time about 2040-50 is projected and a constant steep dicline after. 

## __3__: Comparison to investments in wind and solar electricity production

```python
inv_2 = 'Investment|Energy Supply|Electricity|Solar'
inv_3 = 'Investment|Energy Supply|Electricity|Wind'

df_asia = pyam.read_iiasa(
    'iamc15',
    model='MESSAGEix*',
    region= 'R5ASIA', variable=[inv_2, inv_3], scenario=scens)

df_asia.plot(legend=dict(loc='center left',bbox_to_anchor=(1.0, 0.5)))
```

In contrast to the calculated investments in fossil fuel extraction, in the investments in wind and solar, general rising numbers are projected which again proves the reported key characteristica in the _SR15_ report.
